```python
import mysql.connector

def obter_dados_do_banco(query):
    try:
        conexao = mysql.connector.connect(
            host="127.0.0.1",
            user="root",
            password="",
            database="spotify"
        )
        cursor = conexao.cursor()
        cursor.execute(query)
        resultados = cursor.fetchall()
        return resultados
    except mysql.connector.Error as erro:
        print(f"Erro ao conectar ao MySQL: {erro}")
        return None
    finally:
        if 'conexao' in locals() and conexao.is_connected():
            cursor.close()
            conexao.close()

query_musicas = "SELECT * FROM track_data_final WHERE track_popularity > 50"
dados_filtrados = obter_dados_do_banco(query_musicas)

if dados_filtrados:
    for produto in dados_filtrados:
        print(produto)
```

```python
import math
import pandas as pd
import numpy as np

pedidos_df = pd.read_csv("../Projeto/track_data_final.csv")

# Selecionar os dados de interesse
calculo_popularidade = pedidos_df['track_popularity']

# Cálculo das medidas
media_popularidade = calculo_popularidade.mean()
# Usando np.var com ddof=0 para calcular a variância populacional
variancia_popularidade = np.var(calculo_popularidade, ddof=0)
# Usando np.std com ddof=0 para calcular o desvio padrão populacional
desvio_padrao_popularidade = np.std(calculo_popularidade, ddof=0)

# Coeficiente de Variação (CV)
cv_vendas = (desvio_padrao_popularidade / media_popularidade) * 100

# Distância da Variância em relação à Média
distancia_popularidade = variancia_popularidade / (media_popularidade ** 2)

print(f"--- Análise dos Valores Totais dos Pedidos ---")
print(f"\nMédia dos Valores Totais: {media_popularidade:.2f}")
print(f"Variância dos Valores Totais: {variancia_popularidade:.2f}")
print(f"Desvio Padrão dos Valores Totais: {desvio_padrao_popularidade:.2f}")
print("-" * 100)
print(f"Coeficiente de Variação (CV): {cv_vendas:.2f}%")
print(f"Distância da Variância / Média²: {distancia_popularidade:.2f}")

# Análise de Dispersão
if distancia_popularidade <= 0.10:
    analise_popularidade = "Baixa dispersão dos dados em relação à média."
elif distancia_popularidade < 0.25:
    analise_popularidade = "Dispersão moderada dos dados em relação à média."
else:
    analise_popularidade = "Alta dispersão dos dados em relação à média."
    
print(f"Conclusão da Dispersão: {analise_popularidade}")
```
