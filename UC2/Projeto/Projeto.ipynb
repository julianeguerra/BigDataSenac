***Query para ver as músicas do top 10***

```python
import mysql.connector

# 1. Conectar ao banco de dados
conexao = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="",
    database="spotify"
)

cursor = conexao.cursor()

query_top = """SELECT track_name, track_artist, track_popularity
FROM high_popularity_spotify_data
ORDER BY track_popularity DESC LIMIT 10"""

cursor.execute(query_top)

resultados_top = cursor.fetchall()

for linha in resultados_top:
    print(linha)

cursor.close()
conexao.close()
```

Query com JOIN para ver a relação popularidade X duração

```python
import mysql.connector

# 1. Conectar ao banco de dados
conexao = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="",
    database="spotify"
)

cursor = conexao.cursor()

query_popularidade_duracao = """
SELECT DISTINCT
    h.track_name, 
    h.track_artist, 
    h.track_popularity as popularidade_alta,
    l.track_popularity as popularidade_baixa,
    h.duration_ms as duracao_alta,
    l.duration_ms as duracao_baixa
FROM high_popularity_spotify_data h
JOIN low_popularity_spotify_data l 
    ON h.track_id = l.track_id
ORDER BY h.duration_ms DESC
LIMIT 20
"""

cursor.execute(query_popularidade_duracao)

resultados_pop_duracao = cursor.fetchall()

for linha in resultados_pop_duracao:
    print(linha)

cursor.close()
conexao.close()
```

Query com JOIN para avaliar a relação dançabilidade X popularidade

```python
import mysql.connector

# 1. Conectar ao banco de dados
conexao = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="",
    database="spotify"
)

cursor = conexao.cursor()

query_danca_popularidade = """
SELECT DISTINCT
    h.track_name, 
    h.track_artist, 
    h.track_popularity as popularidade_alta,
    l.track_popularity as popularidade_baixa,
    h.danceability as duracao_alta,
    l.danceability as duracao_baixa
FROM high_popularity_spotify_data h
JOIN low_popularity_spotify_data l 
    ON h.track_id = l.track_id
ORDER BY h.danceability DESC
LIMIT 20
"""

cursor.execute(query_danca_popularidade)

resultados_danca_pop = cursor.fetchall()

for linha in resultados_danca_pop:
    print(linha)

cursor.close()
conexao.close()
```
